Buatlah koding python dengan matplotlib atau seaborn, visualisasi data untuk menampilkan :

Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan grafik histogram (disini saya pecah menjadi 3 grafik)

Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi

Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi

Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Membaca data dari file CSV
data = pd.read_csv('stock.csv')

# Visualisasi distribusi data Purchase Price
plt.figure(figsize=(8, 6))
plt.hist(data['Purchase Price'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Distribution of Purchase Price')
plt.show()

# Visualisasi distribusi data Current Price
plt.figure(figsize=(8, 6))
plt.hist(data['Current Price'], bins=20, color='green', alpha=0.7)
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Distribution of Current Price')
plt.show()

# Visualisasi distribusi data Gain/Loss
plt.figure(figsize=(8, 6))
plt.hist(data['Gain/Loss'], bins=20, color='orange', alpha=0.7)
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Distribution of Gain/Loss')
plt.show()

# Mengambil 5 data dengan Dividend/Share tertinggi
top_dividend = data.nlargest(5, 'Dividend/share')
top_dividend_names = top_dividend[['Stock Name', 'Symbol']]
print('Stocks with Highest Dividend/Share:')
print(top_dividend_names)

# Mengatur interval sumbu x
interval = np.arange(0.2, 2.6, 0.2)

# Visualisasi distribusi Dividend/Share
plt.figure(figsize=(8, 6))
plt.hist(data['Dividend/share'], bins=interval, color='purple', alpha=0.7)
plt.xlabel('Dividend/Share')
plt.ylabel('Frequency')
plt.title('Distribution of Dividend/Share')
plt.xticks(interval)  # Mengatur label sumbu x
plt.show()

# Mengambil 5 data dengan Market Value tertinggi
top_market_value = data.nlargest(5, 'Market Value')
top_market_value_names = top_market_value[['Stock Name', 'Symbol']]
print('Stocks with Highest Market Value:')
print(top_market_value_names)

# Visualisasi distribusi Market Value
plt.figure(figsize=(8, 6))
plt.hist(data['Market Value'], bins=20, color='red', alpha=0.7)
plt.xlabel('Market Value')
plt.ylabel('Frequency')
plt.title('Distribution of Market Value')
plt.show()

# Mengambil 7 data dengan Annual Yield tertinggi
top_annual_yield = data.nlargest(7, 'Annual Yield')
top_annual_yield_names = top_annual_yield[['Stock Name', 'Symbol']]
print('Stocks with Highest Annual Yield:')
print(top_annual_yield_names)

# Visualisasi distribusi Annual Yield
plt.figure(figsize=(8, 6))
plt.hist(data['Annual Yield'], bins=20, color='yellow', alpha=0.7, edgecolor='black')
plt.xlabel('Annual Yield')
plt.ylabel('Frequency')
plt.title('Distribution of Annual Yield')
plt.show()
```
